# Lecture 12: Distribution-free Confidence Intervals

从一个示例开始。

一个随机样本$X_1, X_2, \ldots, X_n$来自未知分布，将区间$(X_{(1)}, X_{(n)})$作为总体中位数$m$的置信区间，计算置信系数。

$$ \text{Sol. }\mathbb{P}(X_{(1)} < m < X_{(n)}) = 1 - \mathbb{P}(m < X_{(1)}) - \mathbb{P}(X_{(n)} < m)
= 1 - \left(\frac{1}{2}\right)^n - \left(\frac{1}{2}\right)^n = 1 - \left(\frac{1}{2}\right)^{n-1}
$$

扩展：一个随机样本 $X_1, \dots, X_n \overset{i.i.d.}{\sim} F$， 对于某个未知的连续分布 $F$，其中位数为 $m$， 那么区间 $(X_{(i)}, X_{(j)})，i < j$ 的置信水平为 $$ \Pr(X_{(i)} < m < X_{(j)}) $$ $$ = \Pr(\text{at least } i \text{ and at most } j-1 \text{ of } X_1, \dots, X_n < m) $$ $$ = \Pr(i \le W \le j-1) = \sum_{k=i}^{j-1} \binom{n}{k} \left(\frac{1}{2}\right)^n $$ 其中，$W$ 表示小于 $m$ 的 $X_i$ 的数量： $$ W = \sum_{i=1}^n \mathbf{1}_{X_i < m} \sim \text{Bin}(n, 1/2). $$

再扩展：对于任意分布 $F$， 设 $q_\alpha$ 为 $F$ 的 $\alpha$ 分位数， 即 $F(q_\alpha) = \alpha$。 那么区间 $(X_{(i)}, X_{(j)})，i < j$ 的置信水平为 $$ \Pr(X_{(i)} < q_\alpha < X_{(j)}) $$ $$ = \Pr(\text{at least } i \text{ and at most } j-1 \text{ of } X_1, \dots, X_n < q_\alpha) $$ $$ = \Pr(i \le W \le j-1) = \sum_{k=i}^{j-1} \binom{n}{k} \alpha^k (1-\alpha)^{n-k} $$ 其中，$W$ 表示小于 $q_\alpha$ 的 $X_i$ 的数量： $$ W = \sum_{i=1}^n \mathbf{1}_{X_i < q_\alpha} \sim \text{Bin}(n, \alpha). $$

## Approximation Using CLT

当 $n$ 较大时，利用中心极限定理， $W$ 可近似为正态分布： $$ W \approx N\left(\frac{n}{2}, \frac{n}{4}\right). $$ 于是，置信水平可近似为 $$ \Pr(X_{(i)} < m < X_{(j)}) \approx \Pr\left(i \le W \le j-1\right) $$ $$ \approx \Phi\left(\frac{j - 1 - n/2}{\sqrt{n}/2}\right) - \Phi\left(\frac{i - n/2}{\sqrt{n}/2}\right). $$ 例如，取 $i = 1, j = n$，则置信水平近似为 $$ \Pr(X_{(1)} < m < X_{(n)}) \approx \Phi\left(\frac{n - 1 - n/2}{\sqrt{n}/2}\right) - \Phi\left(\frac{1 - n/2}{\sqrt{n}/2}\right) $$ $$ = 2\Phi\left(\sqrt{n} - \frac{1}{\sqrt{n}}\right) - 1. $$ 当 $n$ 很大时，置信水平趋近于 $1$。

扩展到其它分位数：对于 $0 < p < 1$，总体 $p$ 分位数为 $\pi_p = F^{-1}(p)$， 则区间 $(X_{(i)}, X_{(j)})$ 的置信水平为 $$ \Pr(X_{(i)} < \pi_p < X_{(j)}) = \Pr(\text{at least } i \text{ and at most } j-1 \text{ of } X_1, \dots, X_n < \xi_p) $$ $$ = \Pr(i \le W_p \le j-1) = \sum_{k=i}^{j-1} \binom{n}{k} p^k (1 - p)^{n-k} $$ 其中，$W_p$ 表示小于 $\pi_p$ 的 $X_i$ 的数量： $$ W_p = \sum_{i=1}^n \mathbf{1}_{X_i < \pi_p} \sim \text{Bin}(n, p). $$ 当 $n$ 较大时，利用中心极限定理， $W_p$ 可近似为正态分布： $$ W_p \approx N(np, np(1 - p)). $$ 于是，置信水平可近似为 $$ \Pr(X_{(i)} < \pi_p < X_{(j)}) \approx \Pr\left(i \le W_p \le j-1\right) $$ $$ \approx \Phi\left(\frac{j - 1 - np}{\sqrt{np(1 - p)}}\right) - \Phi\left(\frac{i - np}{\sqrt{np(1 - p)}}\right). $$  使用半单位修正，得到 $$ \Pr(X_{(i)} < \pi_p < X_{(j)}) \approx \Phi\left(\frac{j - 1 + 0.5 - np}{\sqrt{np(1 - p)}}\right) - \Phi\left(\frac{i - 0.5 - np}{\sqrt{np(1 - p)}}\right). $$
