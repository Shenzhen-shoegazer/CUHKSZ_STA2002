# Lecture 11: Quantile and Q-Q Plot

## Quantile 

分布的 $p$-th 分位数或 $100p$-th 百分位数是具有累积分布函数 $F$ 的值 $\pi_p$，使得 $$ F(\pi_p) = \text{Pr}(X \le \pi_p : X \sim F) = p $$ 例如，当 $p = 1/2$ 时，对应的 $1/2$-th 分位数 $\pi_{1/2}$ 也称为分布 $F$ 的 **中位数**。 当 $p = 1/4$ 或 $3/4$ 时，对应的 $1/4$-th 分位数 $\pi_{1/4}$ 或 $3/4$-th 分位数 $\pi_{3/4}$ 分别称为分布 $F$ 的 **第一** 或 **第三** 四分位数。 

特别地，临界值 $z_\alpha$ 是标准正态分布的 $(1 - \alpha)$-th 分位数。

下面将讨论如何在分布未知的情况下估计分位数$\pi_p$。

### Sample Quantiles 样本分位数

设 $X_1, X_2, \ldots, X_n$ 是来自分布 $F$ 的独立同分布随机变量，记其次序统计量为 $X_{(1)} \leq X_{(2)} \leq \ldots \leq X_{(n)}$，即将样本按从小到大排序后的值。则第 $k$ 个次序统计量 $X_{(k)}$ 称为样本的第 $k$ 个 **样本分位数**。

由于$F(X_{(k)})$服从Beta分布$Beta(k, n-k+1)$，因此$F(X_{(k)})$的期望为$\frac{k}{n+1}$。所以当$n$较大时，$X_{(k)}$可以作为分布$F$的$\frac{k}{n+1}$-th分位数的一个无偏估计量。

但有时，$p$会处于$\frac{k}{n+1}$与$\frac{k+1}{n+1}$之间，这时可以通过插值法来估计分位数$\pi_p$。具体地，设$k$为满足$\frac{k}{n+1} \leq p \leq \frac{k+1}{n+1}$的整数，则$p$-th分位数$\pi_p$的插值估计量为

$$\hat{\pi}_p = x_{(a)} + (p(n+1) - a)(x_{(a+1)} - x_{(a)})$$

其中 $a = \lfloor p(n+1) \rfloor$。这确保了$\hat{\pi}_p$在$x_{(a)}$与$x_{(a+1)}$之间线性变化。

一种特殊情况：当样本量为奇数时，中位数（即$\pi_{1/2}$）的估计量为样本中间的次序统计量；当样本量为偶数时，中位数的估计量为样本中间两个次序统计量的平均值。

形象地说，可以看作从0到1的线段上均匀地取n个点（头尾两个端点不取），将线段分成n+1段，然后根据p落在哪一段来决定分位数的估计值。  
当$p$无限接近0或1时，分位数的估计值会变得不稳定，因此在实际应用中，通常建议选择$p$在$[\frac{1}{n+1}, \frac{n}{n+1}]$范围内。

估计量$\hat{\pi}_p$并不是无偏的，因为无法保证$E[\hat{\pi}_p] = \pi_p$。不过，当样本量$n$较大时，$\hat{\pi}_p$是$\pi_p$的一个一致估计量。

## Q-Q Plot

### Motivation & Definition

在实际应用中，我们通常需要判断数据是否服从某个特定的分布（例如正态分布）。Q-Q图（Quantile-Quantile Plot）是一种常用的图形工具，用于比较两个分布的分位数，从而判断数据是否符合某个假设分布。

Q-Q图的横轴表示理论分布的分位数，纵轴表示样本数据的分位数。如果样本数据与理论分布吻合良好，则Q-Q图上的点应接近一条直线。

### Construction of Q-Q Plot

1. **排序样本数据**：将样本数据 $X_1, X_2, \ldots, X_n$ 按从小到大排序，得到次序统计量 $X_{(1)} \leq X_{(2)} \leq \ldots \leq X_{(n)}$。
2. **计算理论分布的分位数**：对于每个 $i = 1, 2, \ldots, n$，计算理论分布 $F$ 的 $\frac{i}{n+1}$-th 分位数 $\pi_{\frac{i}{n+1}}$。
3. **绘制Q-Q图**：在坐标系中绘制点 $(\pi_{\frac{i}{n+1}}, X_{(i)})$，并观察这些点是否接近一条直线。

### Evaluation

如果Q-Q图上的点大致沿着一条直线分布，说明样本数据与理论分布吻合良好；否则，可能存在偏离。通过观察Q-Q图的形状，可以判断数据的分布特性，例如是否存在重尾、偏态等现象。  
一种常用的方法是计算点与拟合直线之间的残差平方和（RSS），残差平方和越小，说明拟合效果越好。
