# Lectures 6 - 9: Hypothesis Testing

## Basic Concepts

- 从一个实例开始：
  - 假设我们有一个袋子，里面有红球和蓝球，我们想知道红球的比例是多少。
  - 我们随机抽取了一些球，发现其中有60%是红球。
  - 现在我们想知道，这个结果是否足够强烈，以至于我们可以断定袋子里红球的比例确实大于50%。
- 零假设和备择假设：
  - 零假设（Null Hypothesis, $H_0$）：袋子里红球的比例等于50%。
  - 备择假设（Alternative Hypothesis, $H_1$）：袋子里红球的比例大于50%。备择假设可以是单尾的（如这里的例子）或双尾的（如红球比例不等于50%），它和零假设是互斥的。
- 简单假设和复合假设：
  - 简单假设 (Simple Hypothesis)：假设中没有未知参数，例如$H_0: p = 0.5$。
  - 复合假设 (Composite Hypothesis)：假设中包含未知参数，例如$H_1: p > 0.5$，这里$p$是红球的比例。
- 拒绝域和检验统计量：
  - 拒绝域 (Rejection Region)：设样本空间为$\mathcal{D}$，其拥有一个子集$\mathcal{C} \subseteq \mathcal{D}$，在假设检验中，如果观测值落在这个区域$\mathcal{C}$内，我们就拒绝零假设。$\mathcal{C}$不是一个由点构成的集合，而是一个区间或多个区间的集合，例如使$\bar{x}$大于某个临界值的区间。
  - 检验统计量 (Test Statistic, $T$)：用于衡量观测数据与零假设之间差异的量，常见的有z检验和t检验。
- 两种错误类型：
  - 第一类错误 (Type I Error)：错误地拒绝了真实的零假设，记为$\alpha = \mathbb{P}(T \in \mathcal{C}; H_0)$，也称为显著性水平 (significance level)。这里的$\alpha$是事先设定的一个小概率值，例如0.05或0.01。显著性水平和置信水平的和为$1$。
  - 第二类错误 (Type II Error)：错误地接受了假的零假设，记为$\beta = \mathbb{P}(T \notin \mathcal{C}; H_1)$。
- 例如，$X_1,…,X_n \sim N(μ,36)$ 且 $n=16$。令检验统计量 $T = \bar{X} \sim N(\mu, 36/n)$。在 $H_0$ 下， $\bar{X} \sim N(50, 36/16)$， 而在 $H_1$ 下， $\bar{X} \sim N(55, 36/16)$。 再令$\mathcal{C} = \{ \bar{x} \ge 53 \}$，则有 $$\alpha = \text{Pr}(\bar{X} \ge 53; H_0) = \text{Pr}\left(Z \ge \frac{53 - 50}{\sqrt{36/16}}\right) = 0.0228$$ $$\beta = \text{Pr}(\bar{X} < 53; H_1) = \text{Pr}\left(Z < \frac{53 - 55}{\sqrt{36/16}}\right) = 0.0913$$
- $p$值 (p-value)：
  - $p$值是指在零假设为真的前提下，观察到的检验统计量值或更极端的值出现的概率。如果$p$值小于预设的显著性水平$\alpha$，我们就拒绝零假设。
  - $p$值判断和显著性水平判断是等价的。证明：如果$p$值 $\leq \alpha$，则$$\text{Pr}(T \in \mathcal{C}; H_0) = \alpha \geq p\text{-value} = \text{Pr}(T \geq t_{obs}; H_0)$$, 即$t_{obs} \in \mathcal{C}$，所以拒绝$H_0$；反之亦然。

## Hypothesis Testing for Normal Mean 正态均值的假设检验

- 已知$X_1,…,X_n \sim N(μ,σ^2)$，其中$σ^2$已知。我们想检验$H_0: μ = μ_0$ vs. $H_1: μ > μ_0$。检验方法包括$p$值法，临界区域法和置信区间法等，这三种方法是等价的。

### Case 1: $σ^2$已知

- 检验统计量：$$T = \frac{\bar{X} - μ_0}{σ/\sqrt{n}} \sim N(0,1)$$
- 得到样本$x_1,…,x_n$后，计算$$z = \frac{\bar{x} - μ_0}{σ/\sqrt{n}}$$ 
- 三种方法分别得到：
  - $p$值：$$p\text{-value} = \text{Pr}(T \ge z; H_0) = \text{Pr}(Z \ge z) \leq \alpha$$
  - 显著性水平$\alpha$下的拒绝条件：$$z \geq z_{\alpha}$$
  - 置信水平$1-\alpha$下的置信区间：$$\mu_0 \notin \left( \bar{x} - z_{\alpha} \frac{σ}{\sqrt{n}}, +\infty \right)$$ 也可以写作$$\bar{x} - z_{\alpha} \frac{σ}{\sqrt{n}} > μ_0$$
- 下图为$z$的分布图，阴影部分为拒绝域，面积为$\alpha$。如果情况是双尾检验，则拒绝域为两端的区域，面积均为$\alpha/2$。
- 如果是双尾检验，三种方法将分别得到：
  - $p$值：$$p\text{-value} = 2 \cdot \text{Pr}(T \ge |z|; H_0) = 2 \cdot \text{Pr}(Z \ge |z|) \leq \alpha$$
  - 显著性水平$\alpha$下的拒绝条件：$$|z| \geq z_{\alpha/2}$$
  - 置信水平$1-\alpha$下的置信区间： $$\mu_0 \notin \left( \bar{x} - z_{\alpha/2} \frac{σ}{\sqrt{n}}, \bar{x} + z_{\alpha/2} \frac{σ}{\sqrt{n}} \right)$$ 也可以写作$$|\bar{x} - μ_0| > z_{\alpha/2} \frac{σ}{\sqrt{n}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 设定显著性水平和对应的临界值
alpha = 0.05
z_alpha = norm.ppf(1 - alpha)  # 右尾临界值，例如 alpha=0.05 得到 1.645

# 生成标准正态分布曲线
x = np.linspace(-3.5, 3.5, 400)
# 使用标准正态密度显式公式，避免类型检查器将 norm.pdf 的参数限制为标量导致的类型报错
y = np.exp(-0.5 * x**2) / np.sqrt(2 * np.pi)

plt.figure(figsize=(7, 3.5))
plt.plot(x, y, color="#00A0E9", linewidth=2)

# 坐标轴及刻度
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)
# 转为 float 以避免类型检查器将 numpy.float64 视为不兼容标量
plt.axvline(float(z_alpha), color="#0080C0", linestyle="--", linewidth=1)

# 阴影部分（右尾）
plt.fill_between(
    x, y, 0,
    where=(x >= z_alpha).tolist(),
    color="#9EC9EB",
    alpha=0.7
)

# 标注 α
# 显式计算标量 pdf，避免类型检查器将 xy 的第一个元素判定为 ndarray
pdf_z = float(np.exp(-0.5 * z_alpha**2) / np.sqrt(2 * np.pi))
plt.annotate(
    r"$\alpha$",
    xy=(float(z_alpha) + 0.25, pdf_z),
    xytext=(float(z_alpha) + 0.6, pdf_z + 0.05),
    arrowprops=dict(arrowstyle="->", color="#404040"),
    fontsize=14
)

# 其它美化
# 转为内置 float 以消除类型检查器关于 numpy.float64 的告警
plt.text(float(z_alpha), -0.015, r"$z_\alpha$", ha="center", va="top", fontsize=12)
plt.text(0, -0.015, "0", ha="center", va="top", fontsize=12)
plt.yticks([])
plt.xticks([])
plt.xlim(-3.5, 3.5)
plt.ylim(0, 0.45)
plt.tight_layout()

plt.show()
```

### Case 2: $σ^2$未知

- 将标准正态分布替换为自由度为$n-1$的$t$分布（由于用样本标准差$s$来估计总体标准差$σ$，而计算$s$时已经使用了样本均值$\bar{x}$，因此损失了1个自由度），并用$S^2$代替$σ^2$，其他步骤与Case 1类似，因此有：$$T = \frac{\bar{X} - μ_0}{S/\sqrt{n}} \sim t_{n-1}$$ 其中$$S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2$$
- 得到样本$x_1,…,x_n$后，计算$$t = \frac{\bar{x} - μ_0}{s/\sqrt{n}}$$
- 三种方法分别得到：
    - $p$值：$$p\text{-value} = \text{Pr}(T \ge t; H_0) = \text{Pr}(t_{n-1} \ge t) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$t \geq t_{\alpha}(n-1)$$
    - 置信水平$1-\alpha$下的置信区间：$$\mu_0 \notin \left( \bar{x} - t_{\alpha}(n-1) \frac{s}{\sqrt{n}}, +\infty \right)$$ 也可以写作$$\bar{x} - t_{\alpha}(n-1) \frac{s}{\sqrt{n}} > μ_0$$

## Test Equality of Two Normal Means 两个正态均值的假设检验

- 已知$X_1,…,X_n \sim N(μ_X,σ_X^2)$和$Y_1,…,Y_m \sim N(μ_Y,σ_Y^2)$. 在$\sigma_X^2$和$\sigma_Y^2$相等或不等的情况下，欲检验$H_0: μ_X = μ_Y$ vs. $H_1: μ_X \neq μ_Y$（或$H_1: μ_X > μ_Y$，$H_1: μ_X < μ_Y$）。根据$\sigma_X^2$和$\sigma_Y^2$是否相等，检验方法有所不同。

### Case 0: $\sigma_X^2, \sigma_Y^2$均已知，使用z检验
- 检验统计量：$$T = \frac{\bar{X} - \bar{Y}}{\sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}} \sim N(0,1)$$
- 得到样本$x_1,…,x_n$和$y_1,…,y_m$后，计算$$z = \frac{\bar{x} - \bar{y}}{\sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}}$$
- 三种方法分别得到：
    - $p$值：$$p\text{-value} = \text{Pr}(T \ge z; H_0) = \text{Pr}(Z \ge z) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$z \geq z_{\alpha}$$
    - 置信水平$1-\alpha$下的置信区间：$$0 = \mu_X - \mu_Y \notin \left( (\bar{x} - \bar{y}) - z_{\alpha} \sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}, +\infty \right)$$
- 对于双尾检验，类似地可以得到：
    - $p$值：$$p\text{-value} = 2 \cdot \text{Pr}(T \ge |z|; H_0) = 2 \cdot \text{Pr}(Z \ge |z|) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$|z| \geq z_{\alpha/2}$$
    - 置信水平$1-\alpha$下的置信区间： $$0 = \mu_X - \mu_Y \notin \left( (\bar{x} - \bar{y}) - z_{\alpha/2} \sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}}, (\bar{x} - \bar{y}) + z_{\alpha/2} \sqrt{\frac{\sigma_X^2}{n} + \frac{\sigma_Y^2}{m}} \right)$$

### Case 1: Pooled t-test ($\sigma_X^2 = \sigma_Y^2$)
- 回顾：合并方差 (Pooled Variance) 的计算公式为$$S_p^2 = \frac{(n-1)S_X^2 + (m-1)S_Y^2}{n+m-2}$$ 即，将两个样本的方差加权平均，权重为各自的自由度。
- 检验统计量：$$T = \frac{\bar{X} - \bar{Y}}{S_p \sqrt{\frac{1}{n} + \frac{1}{m}}} \sim t(n+m-2)$$
- 得到样本$x_1,…,x_n$和$y_1,…,y_m$后，计算$$t = \frac{\bar{x} - \bar{y}}{s_p \sqrt{\frac{1}{n} + \frac{1}{m}}}$$
- 考虑$H_1: \mu_X > \mu_Y$: $p$值可以表示为$$p = \text{Pr}(T > t; H_0) = \text{Pr}(T > t; T \sim t(n+m-2))$$
- 利用上述三种方法可以等价得到拒绝$H_0$的条件：
    - $p$值：$$p\text{-value} = \text{Pr}(T \ge t; H_0) = \text{Pr}(t_{n+m-2} \ge t) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$t \geq t_{\alpha}(n+m-2)$$
    - 置信水平$1-\alpha$下的置信区间：$$0 = \mu_X - \mu_Y \notin \left( (\bar{x} - \bar{y}) - t_{\alpha}(n+m-2) s_p \sqrt{\frac{1}{n} + \frac{1}{m}}, +\infty \right)$$
- 对于双尾检验，类似地可以得到：
    - $p$值：$$p\text{-value} = 2 \cdot \text{Pr}(T \ge |t|; H_0) = 2 \cdot \text{Pr}(t_{n+m-2} \ge |t|) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$|t| \geq t_{\alpha/2}(n+m-2)$$
    - 置信水平$1-\alpha$下的置信区间： $$0 = \mu_X - \mu_Y \notin \left( (\bar{x} - \bar{y}) - t_{\alpha/2}(n+m-2) s_p \sqrt{\frac{1}{n} + \frac{1}{m}}, (\bar{x} - \bar{y}) + t_{\alpha/2}(n+m-2) s_p \sqrt{\frac{1}{n} + \frac{1}{m}} \right)$$

### Case 2: Welch's t-test ($\sigma_X^2 \neq \sigma_Y^2$)

- 检验统计量：$$T = \frac{\bar{X} - \bar{Y}}{\sqrt{\frac{S_X^2}{n} + \frac{S_Y^2}{m}}}$$
- 由于$\sigma_X^2 \neq \sigma_Y^2$，$T$不再服从$t$分布。Welch提出了一种近似方法，使用自由度$r$的$t$分布来近似$T$的分布，其中$$r = \lfloor {\frac{\left(\frac{S_X^2}{n} + \frac{S_Y^2}{m}\right)^2}{\frac{\left(\frac{S_X^2}{n}\right)^2}{n-1} + \frac{\left(\frac{S_Y^2}{m}\right)^2}{m-1}}} \rfloor$$
- 得到样本$x_1,…,x_n$和$y_1,…,y_m$后，计算$$t = \frac{\bar{x} - \bar{y}}{\sqrt{\frac{s_X^2}{n} + \frac{s_Y^2}{m}}}$$
- 考虑$H_1: \mu_X \neq \mu_Y$: $p$值可以表示为$$p = \text{Pr}(|T| > |t|; H_0) = 2\text{Pr}(T > |t|; T \sim t(r))$$
- 使用三种方法等价得到拒绝$H_0$的条件：
    - $p$值：$$p = 2Pr(t(r) > |t|) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$|t| \geq t_{\alpha/2}(r)$$
    - 置信水平$1-\alpha$下的置信区间：$$0 = \mu_X - \mu_Y \notin \left( (\bar{x} - \bar{y}) - t_{\alpha/2}(r) \sqrt{\frac{s_X^2}{n} + \frac{s_Y^2}{m}}, (\bar{x} - \bar{y}) + t_{\alpha/2}(r) \sqrt{\frac{s_X^2}{n} + \frac{s_Y^2}{m}} \right)$$

### Case 3: Paired t-test (配对t检验)

- 适用于成对数据的情况，例如同一组受试者在不同时间点的测量值，或两组受试者在某种条件下的测量值。
- 设$D_i = X_i - Y_i$，则$D_1,…,D_n \sim N(μ_D,σ_D^2)$，其中$μ_D = μ_X - μ_Y$。仅讨论检验$H_0: μ_D = 0$ vs. $H_1: μ_D \neq 0$。
- 考虑配对检验统计量：$$T = \frac{\bar{D}}{S_D/\sqrt{n}} = \frac{\bar{X} - \bar{Y}}{S_D/\sqrt{n}} \sim t(n-1)$$
- 得到样本$x_1,…,x_n$和$y_1,…,y_n$后，计算$$d_i = x_i - y_i, \bar{d} = \frac{1}{n}\sum_{i=1}^n d_i, s_D^2 = \frac{1}{n-1}\sum_{i=1}^n (d_i - \bar{d})^2$$ $$t = \frac{\bar{d}}{s_D/\sqrt{n}} = \frac{\bar{x} - \bar{y}}{s_D/\sqrt{n}}$$
- $p$值可以表示为$$p = \text{Pr}(|T| > |t|; H_0) = 2\text{Pr}(T > |t|; T \sim t(n-1))$$
- 使用三种方法等价得到拒绝$H_0$的条件：
    - $p$值：$$p = 2Pr(t(n-1) > |t|) \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$|t| \geq t_{\alpha/2}(n-1)$$
    - 置信水平$1-\alpha$下的置信区间：$$0 = \mu_D \notin \left( \bar{d} - t_{\alpha/2}(n-1) \frac{s_D}{\sqrt{n}}, \bar{d} + t_{\alpha/2}(n-1) \frac{s_D}{\sqrt{n}} \right)$$

## Test Equality of Normal Variance 正态方差的假设检验

### Test Equality of One Normal Variance 一个正态方差的假设检验

- 已知$X_1,…,X_n \sim N(μ,σ^2)$. 我们想检验$H_0: σ^2 = σ_0^2$ vs. $H_1: σ^2 \neq σ_0^2$.
- 令$$U = \frac{S^2}{σ^2} (n-1) \sim \chi^2(n-1)$$ 其中$S^2$为样本方差。
- 则在假设$H_0$下，检验统计量$$\chi^2 = \frac{S^2}{σ_0^2} (n-1) \sim \chi^2(n-1)$$
- 拒绝域为$$\mathcal{C} = \{ \chi^2 \geq \chi^2_{\alpha}(n-1) \}$$

### Test Equality of Two Normal Variances 两个正态方差的假设检验

- 已知$X_1,…,X_n \sim N(μ_X,σ_X^2)$和$Y_1,…,Y_m \sim N(μ_Y,σ_Y^2)$. 我们想检验$H_0: σ_X^2 = σ_Y^2$ vs. $H_1: σ_X^2 \neq σ_Y^2$.
- 令$$U = \frac{S_X^2}{σ_X^2} (n-1) \sim \chi^2(n-1), V = \frac{S_Y^2}{σ_Y^2} (m-1) \sim \chi^2(m-1)$$ 其中$S_X^2$和$S_Y^2$分别为样本方差。
- 则在假设$H_0$下，检验统计量$$F = \frac{S_X^2/σ_X^2}{S_Y^2/σ_Y^2} = \frac{U/(n-1)}{V/(m-1)} \sim F(n-1,m-1)$$
- 得到样本$x_1,…,x_n$和$y_1,…,y_m$后，计算$$f = \frac{s_X^2}{s_Y^2}$$
- 当比例$f$过小或过大时，即$f \ge F_{\alpha/2}(n-1, m-1)$ 或 $f \le F_{1-\alpha/2}(n-1, m-1)$时，拒绝$H_0$。
  - 注意，$F_{1-\alpha/2}(n-1, m-1) = \frac{1}{F_{\alpha/2}(m-1, n-1)}$
- 当$f > 1$时，$p$值可以表示为$$p = 2\text{Pr}(F \ge f; F \sim F(n-1,m-1))$$
- 当$f < 1$时，$p$值可以表示为$$p = 2\text{Pr}(F \le f; F \sim F(n-1,m-1)) = 2\text{Pr}\left(F \ge \frac{1}{f}; F \sim F(m-1,n-1)\right)$$
- 使用三种方法等价得到拒绝$H_0$的条件：
    - $p$值：$$p \leq \alpha$$
    - 显著性水平$\alpha$下的拒绝条件：$$f \ge F_{\alpha/2}(n-1, m-1) \text{ (当 f > 1) } \text{ or } f \le F_{1-\alpha/2}(n-1, m-1) \text{ (当 f < 1)}$$
    - 置信水平$1-\alpha$下的置信区间：$$1 \notin \left( \frac{s_Y^2}{s_X^2} F_{{1 - \alpha}/2}(n-1,m-1), \frac{s_X^2}{s_Y^2} F_{\alpha/2}(n-1,m-1) \right)$$
      - 这个置信区间和双侧$(1 - \alpha)100\%$置信区间是一致的。

## Hypothesis Testing for Proportions 比例的假设检验

- 已知$X_1,…,X_n \sim \text{Bernoulli}(p)$，其中$p$为成功概率。令$Y=\sum_{i=1}^n X_i \sim \text{Binomial}(n,p)$，则有$\mathbb{E}[Y]=np$，$\text{Var}(Y)=np(1-p)$。回顾：根据中心极限定理，当$n$较大时，$Y$可以近似服从正态分布，即$$Y \approx N(np, np(1-p)), \hat{p} = \frac{Y}{n} \approx N\left(p, \frac{p(1-p)}{n}\right)$$
- 给定零假设$H_0: p = p_0$。检验统计量为$$T = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \sim N(0,1)$$ 通过样本可以得到$$t = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}$$
- 首先，考虑单尾检验$H_1: p > p_0$:
  - $p$值：$$p\text{-value} = \text{Pr}(T \ge t; H_0) = \text{Pr}(Z \ge t) \leq \alpha$$
  - 显著性水平$\alpha$下的拒绝条件：$$t \ge z_{\alpha}$$
  - 置信水平$1-\alpha$下的置信区间：$$\hat{p} \geq p_0 + z_{\alpha} \sqrt{\frac{p_0(1-p_0)}{n}}$$
- 考虑双尾检验$H_1: p \neq p_0$:
  - $p$值：$$p\text{-value} = 2\text{Pr}(T \ge |t|; H_0) = 2\text{Pr}(Z \ge |t|) \leq \alpha$$
  - 显著性水平$\alpha$下的拒绝条件：$$|t| \ge z_{\alpha/2}$$
  - 置信水平$1-\alpha$下的置信区间：$$p_0 \notin \left( \hat{p} - z_{\alpha/2} \sqrt{\frac{p_0(1-p_0)}{n}}, \hat{p} + z_{\alpha/2} \sqrt{\frac{p_0(1-p_0)}{n}} \right)$$

- 现在，考虑两个比例$p_1, p_2$的假设检验。令$X_1,…,X_n \sim \text{Bernoulli}(p_1)$和$Y_1,…,Y_m \sim \text{Bernoulli}(p_2)$，其中$p_1$和$p_2$分别为两组的成功概率。令$Z = \sum_{i=1}^n X_i \sim \text{Binomial}(n,p_1)$和$W = \sum_{j=1}^m Y_j \sim \text{Binomial}(m,p_2)$，则有$\mathbb{E}[Z]=np_1$，$\text{Var}(Z)=np_1(1-p_1)$，$\mathbb{E}[W]=mp_2$，$\text{Var}(W)=mp_2(1-p_2)$。当$n$和$m$较大时，根据中心极限定理，有$$Z \approx N(np_1, np_1(1-p_1)), W \approx N(mp_2, mp_2(1-p_2))$$ $$\hat{p}_1 = \frac{Z}{n} \approx N\left(p_1, \frac{p_1(1-p_1)}{n}\right), \hat{p}_2 = \frac{W}{m} \approx N\left(p_2, \frac{p_2(1-p_2)}{m}\right)$$ 因此，有$$\hat{p}_1 - \hat{p}_2 \approx N\left(p_1 - p_2, \frac{p_1(1-p_1)}{n} + \frac{p_2(1-p_2)}{m}\right)$$
- 给定零假设$H_0: p_1 = p_2$。检验统计量为$$T = \frac{(\hat{p}_1 - \hat{p}_2) - 0}{\sqrt{\frac{p(1-p)}{n} + \frac{p(1-p)}{m}}} \sim N(0,1)$$ 其中$p = \frac{Z + W}{n + m}$为合并比例。通过样本可以得到$$t = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\frac{\hat{p}(1-\hat{p})}{n} + \frac{\hat{p}(1-\hat{p})}{m}}}$$ 其中，$$\hat{p}_1 = \frac{Z}{n}, \hat{p}_2 = \frac{W}{m}, \hat{p} = \frac{Z + W}{n + m} = \frac{n\hat{p}_1 + m\hat{p}_2}{n + m}$$
- 在此，仅考虑双尾检验$H_1: p_1 \neq p_2$:
  - $p$值：$$p\text{-value} = 2\text{Pr}(T \ge |t|; H_0) = 2\text{Pr}(Z \ge |t|) \leq \alpha$$
  - 显著性水平$\alpha$下的拒绝条件：$$|t| \ge z_{\alpha/2}$$
  - 置信水平$1-\alpha$下的置信区间：$$0 = p_1 - p_2 \notin \left( (\hat{p}_1 - \hat{p}_2) - z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n} + \frac{\hat{p}(1-\hat{p})}{m}}, (\hat{p}_1 - \hat{p}_2) + z_{\alpha/2} \sqrt{\frac{\hat{p}(1-\hat{p})}{n} + \frac{\hat{p}(1-\hat{p})}{m}} \right)$$

## Power Function 检验力函数

- 回顾：第二类错误 (Type II Error) 记为$\beta$，即错误地接受了假的零假设。检验力 (Power) 记为$1-\beta$，即正确地拒绝了假的零假设。检验力函数 (Power Function) 定义为$$K(\mu) = \mathbb{P}(T \in \mathcal{C}|\mu) = 1 - \beta(\mu)$$ 其中$T$为检验统计量，$\mathcal{C}$为拒绝域，$\mu$为参数，可以取参数空间内的任何值。
- 当$H_0: \mu = \mu_0$，即在零假设下，$K(\mu_0) = \alpha$。当$\mu \neq \mu_0$时，$K(\mu) = 1 - \beta(\mu)$表示在参数为$\mu$时拒绝零假设的概率，即检验力。
- 如果$H_0: \mu \leq \mu_0, H_1: \mu > \mu_0$，则需要确保对于所有$\mu > \mu_0$，检验力$K(\mu)$都足够大。$K(\mu)$通常是一个单调递增函数，即随着$\mu$的增大，检验力也增大。因此$\alpha$仍然等于$K(\mu_0)$，而$\beta(\mu)$随着$\mu$的增大而减小。
  - 通常，$\alpha$是一个预先设定的值。因此，可以通过检验力函数确定$\alpha$对应的拒绝域$\mathcal{C}$，从而确定检验统计量$T$。
- 例如，假设$X_1,…,X_n \sim N(\mu,σ^2)$，其中$σ^2$已知。欲检验$H_0: \mu \leq \mu_0$ vs. $H_1: \mu > \mu_0$。令检验统计量$$T = \frac{\bar{X} - \mu_0}{σ/\sqrt{n}} \sim N(0,1)$$ 在显著性水平$\alpha$下，拒绝域为$\mathcal{C} = \{ t \geq z_{\alpha} \}$。则检验力函数为$$K(\mu) = \mathbb{P}(T \in \mathcal{C}|\mu) = \mathbb{P}\left( \frac{\bar{X} - \mu_0}{σ/\sqrt{n}} \geq z_{\alpha} | \mu \right) = \mathbb{P}\left( Z \geq z_{\alpha} - \frac{\mu - \mu_0}{σ/\sqrt{n}} \right) = \Phi\left( z_{\alpha} - \frac{\mu - \mu_0}{σ/\sqrt{n}} \right)$$ 这里$Z = \frac{\bar{X} - \mu}{σ/\sqrt{n}} \sim N(0,1)$，$\Phi$为标准正态分布的累积分布函数。
- 检验力的影响因素：
  - 显著性水平$\alpha$：较大的$\alpha$通常会增加检验力，但也会增加第一类错误的风险。
  - 样本量$n$：较大的样本量通常会增加检验力，因为它减少了估计的标准误差。这将同时减小$\alpha$和$\beta$。
  - 效应大小 (Effect Size)：即备择假设与零假设之间的差异。较大的效应大小通常会增加检验力。
  - 数据的变异性：较低的变异性（如较小的标准差）通常会增加检验力，因为它使得效应更容易被检测到。
- 因此，为了同时控制两类错误，通常需要在设计实验时进行样本量计算 (Sample Size Calculation)，以确保在预设的$\alpha$水平下，检验力达到所需的水平（如80%或90%）。
- 对于比例的假设检验，检验力函数的计算类似。假设$X_1,…,X_n \sim \text{Bernoulli}(p)$，欲检验$H_0: p = p_0$ vs. $H_1: p > p_0$。令$$Y = \sum_{i=1}^n X_i \sim \text{Binomial}(n,p)$$ 则有$\mathbb{E}[Y]=np$，$\text{Var}(Y)=np(1-p)$。
  - 当$n$较小时，可以直接使用二项分布计算检验力函数$$K(p) = \mathbb{P}(Y \in \mathcal{C}|p) = \sum_{y \in \mathcal{C}} \binom{n}{y} p^y (1-p)^{n-y}$$ 其中拒绝域$\mathcal{C}$可以通过显著性水平$\alpha$确定，即找到最小的$k$使得$$\sum_{y=k}^n \binom{n}{y} p_0^y (1-p_0)^{n-y} \leq \alpha$$ 则$\mathcal{C} = \{ y | y \geq k \}$。
  - 当$n$较大时，根据中心极限定理，有$$Y \approx N(np, np(1-p)), \hat{p} = \frac{Y}{n} \approx N\left(p, \frac{p(1-p)}{n}\right)$$ 检验统计量为$$T = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \sim N(0,1)$$ 在显著性水平$\alpha$下，拒绝域为$\mathcal{C} = \{ t \geq z_{\alpha} \}$。则检验力函数为$$K(p) = \mathbb{P}(T \in \mathcal{C}|p) = \mathbb{P}\left( \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \geq z_{\alpha} | p \right) = \mathbb{P}\left( Z \geq z_{\alpha} - \frac{p - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \right) = \Phi\left( z_{\alpha} - \frac{p - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \right)$$ 这里$Z = \frac{\hat{p} - p}{\sqrt{\frac{p(1-p)}{n}}} \sim N(0,1)$。

- 例如，假设 $X_1, \dots, X_{20}$ 是独立同分布的伯努利试验，成功概率为 $p$。要检验 $H_0: p = \frac{1}{2}$, $H_1: p < \frac{1}{2}$， 我们使用检验统计量 $Y := \sum_{i=1}^{20} X_i \sim \text{Binomial}(20, p)$ 
  1. 给定临界区域 $C = \{Y \le 6\}$，推导 $K(p)$ 并计算在 $p = 1/4$ 时的第一类错误 $\alpha$ 和第二类错误 $\beta$。
  2. 需要多少样本才能控制 $\alpha = 0.05$ 和 $\beta(1/4) = 0.1$？
- 解：
  1. $$K(p) = \mathbb{P}(Y \in C | p) = \mathbb{P}(Y \le 6 | p) = \sum_{y=0}^{6} \binom{20}{y} p^y (1-p)^{20-y}$$
    $$\alpha = K(1/2) = \mathbb{P}(Y \le 6 | p=1/2) = \sum_{y=0}^{6} \binom{20}{y} (1/2)^{20} = 0.0577$$
    $$\beta(1/4) = 1 - K(1/4) = 1 - \mathbb{P}(Y \le 6 | p=1/4) = 1 - \sum_{y=0}^{6} \binom{20}{y} (1/4)^y (3/4)^{20-y} = 0.2142$$
  2. 令 $n$ 为样本量，临界区域为 $C = \{Y \le k\}$，其中 $k$ 满足 $\sum_{y=0}^{k} \binom{n}{y} (1/2)^n \le 0.05$。由中心极限定理，$Y \approx N(np, np(1-p))$，则有
    $$\alpha = K(1/2) = \mathbb{P}(Y \le k | p=1/2) \approx \Phi\left(\frac{k - n/2}{\sqrt{n/4}}\right) \le 0.05$$
    $$\Rightarrow k \le n/2 - 1.645\sqrt{n}/2$$
    $$\beta(1/4) = 1 - K(1/4) = 1 - \mathbb{P}(Y \le k | p=1/4) \approx 1 - \Phi\left(\frac{k - n/4}{\sqrt{3n}/4}\right) \le 0.1$$
    $$\Rightarrow k \ge n/4 + 1.282\sqrt{3n}/4$$
    联立以上两个不等式，得到
    $$n/4 + 1.282\sqrt{3n}/4 \le n/2 - 1.645\sqrt{n}/2$$
    解得 $n \approx 30.3, k \approx 10.6$。由于$n$为整数，修正为$n = 31$。需注意$k$的取值须使得$Y$为整数，为符合统计惯例，修正为$k = 10.5$或$k = 11.5$。代入计算可知$k = 10.5$时$n$为31，而$k = 11.5$时$n$为33. 为严格确保$\alpha$和$\beta$的要求，最终选择$n = 33$。
