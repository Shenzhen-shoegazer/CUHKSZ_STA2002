## Random Variables and Distribution Functions 随机变量和分布函数
### Support of a Random Variable 随机变量的支持
+ 离散: $S = \{x: Pr_X(x) > 0\}$
+ 连续: $S = \{x: f_X(x) > 0\}$
### PMF & PDF
+ PMF: $P_X(x) = Pr_X(x)$
+ PDF: $f_X(x) = \frac{d}{dx}F_X(x)$
+ PMF和PDF的支持是相同的，并且它们的积分/求和都等于1。
### Expectation 期望
+ Def: $E[X] = \sum_{x \in S} x P_X(x)$ (离散) or $E[X] = \int_{-\infty}^{+\infty} x f_X(x) dx$ (连续)
+ 线性: $E[aX + b] = aE[X] + b$ 对任何常数 $a, b$.
### Variance 方差
+ Def: $Var[X] = E[X^2] - (E[X])^2$
+ 线性: $Var[aX + b] = a^2 Var[X]$ 对任何常数 $a, b$.
+ 如果$X$和$Y$均存在有限方差，则$Var(X + Y) = Var(X) + Var(Y) + 2Cov(X, Y)$，其中$Cov(X, Y) = E[XY] - E[X]E[Y]$。
+ 如果$X$和$Y$独立，则$Cov(X, Y) = 0$，因此$Var(X + Y) = Var(X) + Var(Y)$。
## Discrete Distributions 离散分布
### Binomial Distribution 二项分布
+ Def: 一个随机变量$X$是在$n$次独立的伯努利试验中成功的次数，则$X$服从参数为$n$和$p$的二项分布，记作$X \sim B(n, p)$。
+ PMF: $P_X(k) = \binom{n}{k} p^k (1-p)^{n-k}, k = 0, 1, \ldots, n$.
+ $E[X] = np$; $Var[X] = np(1-p)$.
### Poisson Distribution 泊松分布
+ Def: 如果一个随机变量$X$在单位时间内发生的事件次数服从参数为$\lambda$的泊松分布，记作$X \sim Pois(\lambda)$。
+ PMF: $P_X(k) = \frac{\lambda^k e^{-\lambda}}{k!}, k = 0, 1, 2, \ldots$.
+ $E[X] = \lambda$; $Var[X] = \lambda$.
## Moment Generating Function 动量生成函数
+ Def: 随机变量$X$的动量生成函数为$M_X(t) = E[e^{tX}]$，其中$t$是一个实数。
+ 如果$X$和$Y$是独立的随机变量，则$M_{X+Y}(t) = M_X(t)M_Y(t)$。
+ 如果$M_X(t)$在包含0的开区间上有限，那么$M_X$可以唯一确定$X$的分布。
+ 对于连续的$X$, $M_X(t) = \int_{-\infty}^{+\infty} e^{tx} f_X(x) dx$.
+ 对于离散的$X$，$M_X(t) = \sum_{x \in S} e^{tx} P_X(x)$.
### Moments from the MGF 矩
+ 如果$M_X$在0的邻域内存在且可微，则对于$n \ge 1$，$M_X^{(n)}(0) = \frac{d^n}{dt^n} E[e^{tX}]|_{t=0} = E[X^n]$.
+ $E[X] = M_X'(0)$; $Var[X] = M_X''(0) - (M_X'(0))^2$.
+ Poisson分布的矩：$M_X(t) = e^{\lambda(e^t - 1)}$，求导得到$M_X'(t) = \lambda e^{\lambda(e^t - 1)} e^t$，$M_X''(t) = \lambda e^{\lambda(e^t - 1)} e^{2t} + \lambda^2 e^{\lambda(e^t - 1)} e^{2t}$.  
因此，$E[X] = M_X'(0) = \lambda$，$Var[X] = M_X''(0) - (M_X'(0))^2 = \lambda + \lambda^2 - \lambda^2 = \lambda$.
## Continuous Distributions 连续分布
### Exponential Distribution 指数分布
+ Def: 一个随机变量$X$服从参数为$\lambda$的指数分布，记作$X \sim Exp(\lambda)$。
+ PDF: $f(x) = \frac{1}{\theta} e^{-x/\theta}, \quad x > 0.$
+ $E[X] = \theta$; $Var[X] = \theta^2$.
+ MGF: $M_X(t) = \int_{0}^{+\infty} e^{tx} \frac{1}{\theta} e^{-x/\theta} dx = \frac{1}{1 - \theta t}, t < \frac{1}{\theta}$.
### Gamma Distribution 伽马分布
+ Def: 一个随机变量$X$具有形状参数$\alpha$和尺度参数$\theta$的伽马分布，记作$X \sim Gamma(\alpha, \theta)$。
+ PDF: $f(x) = \frac{1}{\Gamma(\alpha) \theta^\alpha} x^{\alpha-1} e^{-x/\theta}, \quad x > 0.$
+ $E[X] = \alpha \theta$; $Var[X] = \alpha \theta^2$.
+ MGF: $M_X(t) = \int_{0}^{+\infty} e^{tx} \frac{1}{\Gamma(\alpha) \theta^\alpha} x^{\alpha-1} e^{-x/\theta} dx = \frac{1}{(1 - \theta t)^\alpha}, t < \frac{1}{\theta}$.
+ 与指数分布的关系: 如果$X \sim Gamma(1, \theta)$，则$X \sim Exp(\theta)$。
+ 求和：如果$X_1 \sim \text{Gamma}(\alpha_1, \theta), X_2 \sim \text{Gamma}(\alpha_2, \theta)$ $\Rightarrow X_1 + X_2 \sim \text{Gamma}(\alpha_1 + \alpha_2, \theta)$
### Chi-Square Distribution 卡方分布
+ Def: 一个随机变量$X$服从自由度为$r$的卡方分布，记作$X \sim \chi^2(r)$。
+ PDF: $f(x) = \frac{1}{2^{r/2} \Gamma(r/2)} x^{r/2 - 1} e^{-x/2}, \quad x > 0.$
+ $E[X] = r$; $Var[X] = 2r$.
+ MGF: $M_X(t) = (1 - 2t)^{-r/2}, t < \frac{1}{2}$.
### Normal Distribution 正态分布
+ Def: 一个随机变量$X$服从均值为$\mu$，方差为$\sigma^2$的正态分布，记作$X \sim N(\mu, \sigma^2)$。
+ PDF: $f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}, \quad -\infty < x < +\infty.$
+ $E[X] = \mu$; $Var[X] = \sigma^2$.
+ MGF: $M_X(t) = e^{\mu t + \frac{1}{2} \sigma^2 t^2}$.
+ 标准正态分布: $Z = \frac{X - \mu}{\sigma} \sim N(0, 1)$.
+ 如果$X \sim N(\mu, \sigma^2)$，则$Z = \frac{X - \mu}{\sigma} \sim N(0, 1)$; $Z^2$服从卡方分布，记作$Z^2 \sim \chi^2(1)$.
### Student's t-Distribution
+ Def: 令$T = \frac{Z}{\sqrt{U/r}}$，其中$Z \sim N(0, 1)$，$U \sim \chi^2(r)$，$Z$和$U$独立，则$T \sim t(r)$.
+ PDF: $f(t) = \frac{\Gamma\left(\frac{r+1}{2}\right)}{\sqrt{r\pi} \Gamma\left(\frac{r}{2}\right)} \left(1 + \frac{t^2}{r}\right)^{-\frac{r+1}{2}}, \quad -\infty < t < +\infty.$
+ $E[T] = 0$; $Var[T] = \frac{r}{r-2}, r > 2.$
+ MGF: $M_T(t) = \left(1 - 2t\right)^{-r/2}, t < \frac{1}{2}$.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'DejaVu Sans']  # 支持中文显示
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号

# 设置x轴范围
x = np.linspace(-4, 4, 1000)

# 计算自由度为1和30的t分布概率密度函数
t_df1 = stats.t.pdf(x, df=1)
t_df30 = stats.t.pdf(x, df=30)

# 为了比较，也计算标准正态分布
normal = stats.norm.pdf(x, loc=0, scale=1)

# 创建图像
plt.figure(figsize=(8, 5))
plt.plot(x, t_df1, 'r-', linewidth=2, label='t分布 (自由度=1)')
plt.plot(x, t_df30, 'b-', linewidth=2, label='t分布 (自由度=30)')
plt.plot(x, normal, 'g--', linewidth=2, label='标准正态分布 N(0,1)')

# 添加图形装饰
plt.xlabel('t值')
plt.ylabel('概率密度')
plt.title("Student's t分布与标准正态分布的比较")
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-4, 4)

# 显示图形
plt.tight_layout()
plt.show()

# 打印一些统计信息
print("分布特性比较:")
print(f"t分布(df=1): 均值=0, 方差=无限大 (重尾分布)")
print(f"t分布(df=30): 均值=0, 方差={30/(30-2):.3f}")
print(f"标准正态分布: 均值=0, 方差=1")
print("\n观察:")
print("- 自由度=1的t分布具有更重的尾部，峰值更低")
print("- 自由度=30的t分布已经非常接近标准正态分布")
print("- 随着自由度增加，t分布逐渐趋近于标准正态分布")
```

### F Distribution
+ 令$F=\frac{U / r_1}{V / r_2}$，其中$U$和$V$分别服从自由度为$r_1$和$r_2$的卡方分布，且相互独立，那么$F$服从自由度为$(r_1, r_2)$的F分布。
## Two Limit Theorems
1. **中心极限定理 (Central Limit Theorem)**: 对于独立同分布的随机变量序列，当样本容量足够大时，其样本均值的分布趋近于正态分布。
2. **大数法则 (Law of Large Numbers)**: 随着样本容量的增加，样本均值几乎必然收敛于总体均值。
+ 半单位修正 (Half-Unit Correction): 当离散分布被用于连续分布的近似时，进行的调整以提高估计的准确性。
+ Student's Theorem: 如果$X_1, X_2, \ldots, X_n$是来自正态分布$N(\mu, \sigma^2)$的独立样本，且$\bar{X}$和$S^2$分别是样本均值和样本方差，则统计量$T = \frac{\bar{X} - \mu}{S/\sqrt{n}}$服从自由度为$n-1$的t分布。
